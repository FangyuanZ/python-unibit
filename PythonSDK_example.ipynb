## Get real-time price of GOOGLE 

```python
import csv
sp = StockPrice(key=API_KEY)
#return the latest "size" price of GOOGLE
#data tyoe can either be json or csv
google_price = sp.getPricesRealTime("GOOGL", size=50, datatype="json")
#change the format
GOOGL_intra = pd.DataFrame(google_price)
GOOGL_intra.head(20)
```

### ----Visualize the real-time stock price of GOOGLE

```python
GOOGL_intra["time_index"] = pd.to_datetime((GOOGL_intra['date'] + GOOGL_intra["minute"]).values, format='%Y%m%d%H:%M')
GOOGL_intra.set_index("time_index",inplace=True)
GOOGL_intra.tail(40).price.plot(figsize=[8,8], title = 'Visuliaze intra_day stock proce for GOOGLE')
```

## Get historical information(meta data) of GOOGLE

```python
hist = StockPrice(key=API_KEY)
#rng is the range to grab historical prices, either 1m, 3m, 1y, 3y, 5y, 10y, or 20y
#interval is the a positive number (n). If passed, chart data will return every nth element as defined by Interval
google_hist_price = sp.getPricesHistorical("GOOGL", date_range="1y", interval = 10, datatype="json")
# google_hist_price
GOOGL_hist = pd.DataFrame(google_hist_price["Stock price"])
GOOGL_hist.head(20)
```

### ---Visulize the historical stock data(close price and volume)

```python
GOOGL_hist.date = pd.to_datetime(GOOGL_hist.date)
GOOGL_hist.set_index("date",inplace=True)
GOOGL_hist.close.plot(figsize=[8,8], color='red', title = 'Historical close price for GOOGLE')
```

```python
GOOGL_hist.volume.plot(figsize=[8,8], color='green', title = 'Historical volume for GOOGLE')
```
